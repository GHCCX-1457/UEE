```python
import random
import numpy as np
import csv
#写入csv文件
with open('students.csv','w',newline='',encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['chinese','math','english','physics','chemistry','biology','politics','history','geography','location','time','wenke_score','like_score'])

#模拟生成语文成绩
chinese = [random.randint(60,150) for i in range(5000)]
#模拟生成数学成绩
math = [random.randint(30,150) for i in range(5000)]
#模拟生成英语成绩
english = [random.randint(40,150) for i in range(5000)]
#模拟生成物理成绩
physics = [random.randint(30,100) for i in range(5000)]
#模拟生成化学成绩
chemistry = [random.randint(30,100) for i in range(5000)]
#模拟生成生物成绩
biology = [random.randint(30,100) for i in range(5000)]
#模拟生成政治成绩
politics = [random.randint(30,100) for i in range(5000)]
#模拟生成历史成绩
history = [random.randint(30,100) for i in range(5000)]
#模拟生成地理成绩
geography = [random.randint(30,100) for i in range(5000)]
#模拟所有省份
locations = ["北京","天津","上海","重庆","河北","山西","辽宁","吉林","黑龙江","江苏",
"浙江","安徽","福建","江西","山东","河南","湖北","湖南","广东","海南","四川","贵州",
"云南","陕西","甘肃","青海","内蒙古","广西","西藏","宁夏","新疆"]

#模拟生成考试时间
times =["2019","2016","2017","2018","2019","2020","2021","2022","2023","2024","2025"]
students = []

for i in range(len(geography)):
    scores = []
    scores.append(chinese[i])
    scores.append(math[i])
    scores.append(english[i])
    scores.append(physics[i])
    scores.append(chemistry[i])
    scores.append(biology[i])#模拟生成生物成绩
    scores.append(politics[i])#模拟生成政治成绩
    scores.append(history[i])#模拟生成历史成绩
    scores.append(geography[i])#模拟生成地理成绩
    scores.append(locations[random.randint(0,30)])
    scores.append(times[random.randint(0,10)])
    scores.append(chinese[i]+math[i]+english[i]+politics[i]+history[i]+geography[i])
    scores.append(chinese[i]+math[i]+english[i]+physics[i]+chemistry[i]+biology[i])

    students.append(scores)

    with open('students.csv','a+',newline='',encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(scores)
print(students)


```
